```python
#Importamos las librerias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Librerias para modelos de ciencia de Datos
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Leemos los datos para cargarlos en la variable

data = pd.read_csv('./car_details.csv')
```

```python
#1. Analisis Exploratorio de los Datos:

data.head(50)
```

```python
#Descripcion de los datos

data.describe()
```

```python
#2. Preprocesar los datos limpi√°ndolos:

data.describe().apply(lambda s: s.apply(lambda x: format(x, 'g')))
```

```python
# Grafico para identificar valores atipicos

plt.figure(figsize=(6, 3))
sns.boxplot(x=data['selling_price'])
plt.title('Identificaci√≥n Datos Faltantes y At√≠picos', fontsize=10)
```

```python
#Buscamos la relacion entre precio y el a√±o del modelo

plt.scatter(data['year'], data['selling_price'], color = 'lightcoral')
plt.title('Precio vs A√±o')
plt.xlabel('A√±o del Vehiculo')
plt.ylabel('Precio de Venta')
plt.box(False)
plt.show()
```

```python
#Buscamos la relacion entre precio y el Kilometraje

plt.scatter(data['km_driven'], data['selling_price'], color = 'lightcoral')
plt.title('Precio vs Kilometraje')
plt.xlabel('Kilometraje')
plt.ylabel('Precio de Venta')
plt.box(False)
plt.show()
```

```python
#Revisamos Datos Faltantes

print ("Vehiculos sin registro en A√±o de Modelo: ",data['year'].isnull().sum().sum())
print ("Vehiculos sin registro en Kilometraje: ",data['km_driven'].isnull().sum().sum())
print ("Vehiculos sin registro en numeros de sillas: ",data['seats'].isnull().sum().sum())
print ("Vehiculos sin precio de Venta: ",data['selling_price'].isnull().sum().sum())
```

```python
#Reemplazamos los valores nulos con un 0

data['seats'].fillna(0, inplace = True)
```

```python
#Reemplazamos el valor 0 con el promedio de la columna

data['seats']=data['seats'].replace(0,data['seats'].mean())
```

```python
print ("Vehiculos sin registro en numeros de sillas: ",data['seats'].isnull().sum().sum())
data.describe().apply(lambda s: s.apply(lambda x: format(x, 'g')))
```

```python
data.info()
```

```python
#3. Seleccionar las caracter√≠sticas m√°s relevantes para entrenar:
#El objetivo es predecir el precio del autom√≥vil en funci√≥n de sus atributos

data['selling_price'].value_counts()
```

```python
#Dividimos el Dataset entre la variable Y dependiente y la X independiente

X = data[['year']]  # independente
Y = data['selling_price']  # dependente
```

```python
#4. Dividir el dataset en Train y Test para evaluar:
#Usamos 75% para entrenamiento y 25% para pruebas

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
#Se usa la funcion train_test_split para separar el bloque de entrenamiento
print (X_train)
```

```python
print(Y_train)
```

```python
#5. Entrenamos el Modelo

modelo_regresion = LinearRegression()
modelo_regresion.fit(X_train, Y_train)
```

```python
Y_prediccion_test = modelo_regresion.predict(X_test)
Y_prediccion_train = modelo_regresion.predict(X_train)
```

```python
print(Y_prediccion_test)
```

```python
print(Y_prediccion_train)
```

```python
print(Y_test)
```

```python
#6. Evaluar el desempe√±o del modelo en el conjunto de Test:

mse = mean_squared_error(Y_test, Y_prediccion_test)
r2 = r2_score(Y_test, Y_prediccion_test)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
#7. Realizar las diferentes gr√°ficas que permitan visualizar los resultados del modelo:

plt.scatter(X_test, Y_test, color = 'blue')
plt.plot(X_train, Y_prediccion_train, color = 'firebrick')
plt.title('Precio vs Modelo (Pruebas)')
plt.xlabel('Modelo del Vehiculo')
plt.ylabel('Precio del Vehiculo')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Valor/A√±o', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_train, Y_train, color = 'blue')
plt.plot(X_train, Y_prediccion_train, color = 'firebrick')
plt.title('Precio vs Modelo (Entrenamiento)')
plt.xlabel('Modelo del Vehiculo')
plt.ylabel('Precio del Vehiculo')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Valor/A√±o', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

#8. Interpretar, analizar y documentar los resultados obtenidos.

Se pueden interpretar los resultados obtenidos que el precio del veh√≠culo incrementa en relaci√≥n con el a√±o del modelo, de tal manera que el veh√≠culo entre m√°s nuevo o reciente sea el a√±o del modelo va a tener un valor m√°s alto en el mercado.

Por la tanto la grafica nos ayuda a interpretar que es directamente proporcional.
